# || Observations ||
1. The SK_ID_BUREAU(PK) contains recoded id's of previous credits (active and closed) of our original loan id SK_ID_CURR (FK). \
2. The CREDIT_TYPE has 15 nominal values. By using One Hot Encoding, the dimensions of this dataset will be doubled and simply using label encoding can unintensionally suggest order in the values. During FE, some other way is to be used to preserve the dimensionality of the dataset and encode this column.
3. The ANNUITY has more than 70% missing values. Deeper investigation is required to keep this column or drop. However, as the data contains closed loans also, so the annuity for them becomes zero.<br> 
The DAYS_CREDIT_ENDDATE,DAYS_ENDDATE_FACT,AMT_CREDIT_MAX_OVERDUE columns have missing values, but these column provide information subject to a condition. The remaining days of the ongoing CB credit (if there is any), then days since prev CB credit ended (if there was any and it has ended) and max over due amount, if there was any.<br>
The AMT_DEBT shows the remaining debt of the borrowers that have active loan. This feature shows how much a borrower is already under debt. The CC_LIMIT gives information about if a borrower has already utilized the provided limit.
4. 14% of Cash loans have been sold. Further investigation required regarding sample size and overall effect.
